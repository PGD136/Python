딥러닝 기초 수학

```python
import math #수학 패키지
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
# 일차 함수 y = ax + b
def linear_function(x):
    a = 0.6 # 기울기
    b = 2 # y 절편

    return a * x + b

print(linear_function(5))
```

```python
x = np.arange(-5,5,0.1) # -5부터 5까지 0.1값씩 느는 배열 생성
y = linear_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
# 이차함수 y = ax2 + bx + c

def quadratic_function(x):
    a = 1
    b = -1
    c = -2

    return a*x**2 + b*x + c

print(quadratic_function(2))
```

```python
y = quadratic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python
# 삼차함수(다항함수) y = ax3 + bx2 + cx + d (숫자는 제곱임 **제곱)

def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8

    return a*x**3 + b*x**2 + c*x + d
```

```python
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
```

```python
# 함수의 최소값 / 최대값
def my_func(x):
    a = 1
    b = -3
    c = 10

    return a*x**2 + b*x + c #이차 함수
```

```python
x = np.arange(-10, 10, 0.1)
y = my_func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My Function')

plt.scatter(1.5, my_func(1.5)) #원하는 지점에 점 찍기(최소점)
```

```python
# 지수함수 y = ax

def exponential_function(x):
    a = 4
    return a**x
```

```python
print(exponential_function(4))
print(exponential_function(0))
```

```python
x = np.arange(-3, 2, 0.1)
y = exponential_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-4, 3)
plt.ylim(-1, 15)
plt.title('Exponetial Function')
```
