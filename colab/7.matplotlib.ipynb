```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x))

```

```python
plt.plot(x,np.sin(x-0), color='blue')          #색상의 이름으로 지정
plt.plot(x,np.sin(x-1), color='k')             #색상 약어 g,r,k,b 등 사용
plt.plot(x,np.sin(x-2), color='0.5')           #0~1 사이에 회색조를 표현한다. 0일수록 검정(빛의 밝기)
plt.plot(x,np.sin(x-3), color='#FFDD44')       #16진수로 표현한 RGB
plt.plot(x,np.sin(x-4), color=(1.0,0.2,0.3))   #RGB 값을 튜플로 표현
plt.plot(x,np.sin(x-5), color='chartreuse')    #색의 이름으로 지정
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g') #줄임으로 사용할 시 약어만 가능
plt.plot(x, x+1, 'r-')
plt.plot(x, x+2, '--c')
plt.plot(x, x+2, '-.k')
```

```python
#플롯 조정하기: 축 경계
plt.plot(x,np.sin(x))
plt.xlim(-1,11)      # x축 보이는 지점 정하기(시작,끝)
plt.ylim(-1.5,1.5)   # y축 보이는 지점 정하기(시작,끝)
#plt.axis([-1,11,-1.5,1.5]) 한번에 사용가능하지만 잘 사용 안함
```

```python
plt.plot(x,np.sin(x))
plt.xlim(10,0)       # x축 반전
plt.ylim(1.2,-1.2)   # y축 반전
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight') # 타이트하게 붙어서 나온다
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal') #같은사이즈 같은단위 적용(현실에 가까운 그래프)
```

```python
#플롯에 레이블 붙이기(꾸미기)
plt.plot(x,np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x,np.sin(x),'-g',label='sin(x)') # 선에 라벨작업
plt.plot(x,np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend() #선에 라벨 보여줌
```

```python
# 산점도의 출력
x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x, y, 'd', color='black')
```

```python
rng = np.random.RandomState(0)#시드값0
for marker in ['o', ',', 'x', '+', 'v', '<', '>', 's', 'd', 'p']:   #마크의 종류
    plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
    plt.legend()
    plt.xlim(0,1.8)
```
